```
#Reading the data
first_file = pd.read_excel('/content/drive/MyDrive/Study/Self Study/MKR Data Analysis/2. Internship First Task - HIV/HIV_0000000006.xlsx')
second_file = pd.read_excel('/content/drive/MyDrive/Study/Self Study/MKR Data Analysis/2. Internship First Task - HIV/HIV_0000000026,SDGHIV (1).xlsx')
third_file = pd.read_excel('/content/drive/MyDrive/Study/Self Study/MKR Data Analysis/2. Internship First Task - HIV/HIV_0000000001 (2).xlsx')
```

```
print("Type of 'Number of deaths due to HIV/AIDS' : ", type(first_file['Number of deaths due to HIV/AIDS'][0]))
# print("Type of 'year' : ", type(first_file['Year'][0]))
#Here we can se tha datatype of "Type of 'Number of deaths due to HIV/AIDS" column are string
#So first we need to remove values within brackets and then convert into integer, 
#so that we can use these numeric values effeciently

#Also we can convert datatype of year to year object to use it more efficiently
```

```
def remove_brackets_value(value):
    # https://stackoverflow.com/questions/640001/how-can-i-remove-text-within-parentheses-with-a-regex  --- Reference for below line
    value = re.sub(r'\[[^]]*\]', '', value)
    value = re.sub(r'\s', "", value)
    value = re.sub(r'[^0-9.]+',"", value)
    return float(value)
```

```
#Let Remove Values within Brackets
first_file['Number of deaths due to HIV/AIDS'] = first_file['Number of deaths due to HIV/AIDS'].apply(remove_brackets_value)

#Now converting Year into datetime object
# first_file['Year'] = pd.to_datetime(first_file['Year'], format= '%Y') 
```

```
first_file.to_csv("first_file_HIV_cleaned.csv", index = False)
```

```
#Let's Checkout the Null values
second_file.isnull().sum()
# There are no null values in any column
```

```
#Removing The First Row
second_file = second_file[1:]

#Renaming the First and Second Column name as well
#New_Case_Per_1000_Male_Female - (New HIV infections (per 1000 uninfected population) - Both Sex)
#New_Case_Per_1000_Male -        (New HIV infections (per 1000 uninfected population) - Male)
#New_Case_Per_1000_Female -        (New HIV infections (per 1000 uninfected population) - Female)

second_file.rename(columns= {'Unnamed: 0' : 'WHO Region', 'Unnamed: 1':'Year',
                             'Number of new HIV infections' : 'New_Cases',
                             'New HIV infections (per 1000 uninfected population)' : 'New_Case_Per_1000_Male_Female',
                             'New HIV infections (per 1000 uninfected population).1' : 'New_Case_Per_1000_Male',
                             'New HIV infections (per 1000 uninfected population).2' : 'New_Case_Per_1000_Female' 
                            }, inplace = True)
```

```
#Now converting Year into datetime object
# second_file['Year'] = pd.to_datetime(second_file['Year'], format= '%Y') 

#Remvoing Values withing brackets and unwanted characters
second_file['New_Cases'] =  second_file['New_Cases'].apply(remove_brackets_value)
second_file['New_Case_Per_1000_Male_Female'] =  second_file['New_Case_Per_1000_Male_Female'].apply(remove_brackets_value)
second_file['New_Case_Per_1000_Male'] =  second_file['New_Case_Per_1000_Male'].apply(remove_brackets_value)
second_file['New_Case_Per_1000_Female'] =  second_file['New_Case_Per_1000_Female'].apply(remove_brackets_value)
```
